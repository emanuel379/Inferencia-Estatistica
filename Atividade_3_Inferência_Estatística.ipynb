**a)** Calcule o tamanho da amostra que deve ser coletada para assegurar um erro máximo de 2% na estimação da verdadeira proporção de eleitores favoráveis a um determinado candidato. Considere α = 5%.

```r
# Parâmetros de entrada
e = 0.02  # Erro máximo permitido
p = 0.5   # Proporção esperada
N = 124252796  # Tamanho da população
level = 0.95  # Nível de confiança
```

```r
# Cálculo do tamanho da amostra
z = qnorm((1 + level) / 2)  # Valor crítico para o nível de confiança
n = ceiling((z^2 * p * (1 - p)) / (e^2))
cat("O tamanho de amostra que deve ser coletada é:", n, "\n")
```

**b)** Considere as informações populacionais de votos dos eleitores, disponível no arquivo POPULACAO.txt. Colete uma amostra de tamanho n (obtido no item a)considerando os seguintes comandos no R:
image.png

em que k é um número conhecido de identificação. Considerando uma margem de erro de 2% e baseada na amostra coletada, obtenha a intenção de voto para cada candidado (C1 e C2), Brancos (B) e Nulos(N), juntamente com os seus respectivos intervalos (otimista e conservativo) com 95% de confiança.

```r
# Fixar a semente
dados=read.table("/content/POPULACAO.txt") # Leitura do arquivo POPULACAO.txt
n= 2401 # Calculado no item a
k = 3
set.seed(k)
i=sample(1:n,n,F)
amostra=dados[i,]
amostra
```

```r
# Calculando as proporções amostrais para cada categoria de voto

prop_c1 = sum(amostra == "C1")/n
prop_c2 = sum(amostra == "C2")/n
prop_nulo= sum(amostra == "N")/n
prop_branco = sum(amostra == "B")/n
```

```r
# Calculando os intervalos de confiança para cada proporção

# Fixado o erro de 2%
# Abordagem otimista

IC_c1_otimista_inf = prop_c1 - 0.02
IC_c1_otimista_sup = prop_c1 + 0.02

IC_c2_otimista_inf = prop_c2 - 0.02
IC_c2_otimista_sup = prop_c2 + 0.02

IC_nulo_otimista_inf = prop_nulo - 0.02
IC_nulo_otimista_sup= prop_nulo + 0.02

IC_branco_otimista_inf = prop_branco - 0.02
IC_branco_otimista_sup = prop_branco + 0.02

# Abordagem conservativa

IC_c1_conservativa_inf = prop_c1 - 0.02
IC_c1_conservativa_sup = prop_c1 + 0.02

IC_c2_conservativa_inf = prop_c2 - 0.02
IC_c2_conservativa_sup = prop_c2 + 0.02

IC_nulo_conservativa_inf = prop_nulo - 0.02
IC_nulo_conservativa_sup = prop_nulo + 0.02

IC_branco_conservativa_inf = prop_branco - 0.02
IC_branco_conservativa_sup = prop_branco + 0.02
```

# **Resultados**

```r
# Intervalo de confiança para C1

cat("Intenção de voto para C1: [",prop_c1,"]", "\n")
cat('IC para C1 abordagem otimista:[', IC_c1_otimista_inf, IC_c1_otimista_sup, "]", "\n")
cat('IC para C1 abordagem conservativa:[', IC_c1_conservativa_inf, IC_c1_conservativa_sup,"]")
```

```r
# Intervalo de confiança para C2

cat("Intenção de voto para C1: [",prop_c2,"]", "\n")
cat('IC para C2 abordagem otimista:[', IC_c2_otimista_inf, IC_c2_otimista_sup, "]", "\n")
cat('IC para C2 abordagem conservativa:[', IC_c2_conservativa_inf, IC_c2_conservativa_sup,"]")
```

```r
# Intervalo de confiança para Nulos

cat("Intenção de voto para nulo: [",prop_nulo,"]", "\n")
cat('IC nulo abordagem otimista: [', IC_nulo_otimista_inf, IC_nulo_otimista_sup, "]", "\n")
cat('IC nulo abordagem conservativa: [', IC_nulo_conservativa_inf, IC_nulo_conservativa_sup, "]")
```

```r
# Intervalo de confiança para Brancos

cat("Intenção de voto para branco: [",prop_branco,"]", "\n")
cat('IC branco abordagem otimista: [', IC_branco_otimista_inf, IC_branco_otimista_sup,"]", "\n")
cat('IC branco abordagem conservativa: [', IC_branco_conservativa_inf, IC_branco_conservativa_sup,"]")
```
